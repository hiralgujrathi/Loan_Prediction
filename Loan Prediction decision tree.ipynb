## Convert the dataset to numerical form

```python
from sklearn import preprocessing
#sklearn.preprocessing package provides several common utility functions and transformer classes to change raw feature vectors into a representation that is more suitable and understandable form
```

```python
#fit transform=Fit label encoder and return encoded labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
#fit transform=Fit label encoder and return encoded labels.
#Label Encoding is a technique that is used to convert categorical columns into numerical ones so that they can be fitted by machine learning models which only take numerical data.
Train['Gender']= label_encoder.fit_transform(Train['Gender'])
Train['Married']= label_encoder.fit_transform(Train['Married'])
Train['Education']= label_encoder.fit_transform(Train['Education'])
Train['Self_Employed']= label_encoder.fit_transform(Train['Self_Employed'])
Train['Property_Area']= label_encoder.fit_transform(Train['Property_Area'])
Train['Loan_Status']= label_encoder.fit_transform(Train['Loan_Status'])
```
